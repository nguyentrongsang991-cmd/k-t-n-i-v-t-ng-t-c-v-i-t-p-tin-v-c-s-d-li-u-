# # Lưu thành file CSV 

```python
 # Lưu thành file CSV 
df.to_csv("VCI_data.csv", index=False, encoding="utf-8-sig")
print ('Đã lưu VCI_data.csv')
```

```python
# Đọc file CSV 
df = pd.read_csv("VCI_data.csv") # VCI_data.csv tên file CSV
```

```python
df
```

```python
# Xóa cột không cần thiết
df = df.drop(columns=["open", "high", "low",])
```

```python
df
```

```python
# tính returns cùng độ dài với df (hàng đầu tiên sẽ là NaN)
df['returns'] = pd.to_numeric(df['close'], errors='coerce').pct_change()

# loại bỏ hàng NaN đầu tiên:
df = df.dropna(subset=['returns']).copy()
```

```python
df
```

```python
df.info()
```

```python
# 1) Chuẩn hóa cột thời gian về kiểu datetime
df['time'] = pd.to_datetime(df['time'], errors='coerce')

# 2) Loại bỏ dòng lỗi thời gian (nếu có) và trùng lặp theo time
df = df.dropna(subset=['time']).drop_duplicates(subset=['time'])

# 3) Đặt index là time và sắp xếp tăng dần theo thời gian
df = df.set_index('time').sort_index()
df.index.name = 'time'   # tuỳ chọn: đặt tên rõ ràng cho index

print(df.head())
print(df.index.dtype)    # kiểm tra kiểu datetime64[ns]
```

```python
df.describe() # thông kê mô tả
```

```python
# Vẽ đồ thị giá đóng cửa
df['close'].viz.timeseries(figsize=(10, 6),
		title='Giá đóng cửa - VCI',
		ylabel='Giá',
		xlabel='Thời gian',
		color_palette='vnstock',
		palette_shuffle=True)
```

```python
# Vẽ đồ thị xu hướng giá đóng của 
plt.figure(figsize=(10, 6))                 # <- đặt kích thước ở đây
sns.lineplot(x=df.index, y=df['close'])     # truyền cả x,y là vector
plt.title("Xu hướng giá cổ phiếu VIC")
plt.xlabel("Ngày")
plt.ylabel("Giá đóng cửa (VND)")
plt.gcf().autofmt_xdate()
plt.tight_layout()
plt.show()
```

```python
# vẽ biểu đồ phân phối (histogram) của lợi suất hàng ngày cổ phiếu VIC
sns.histplot(df["returns"].dropna(), kde=True)
plt.title("Phân phối lợi suất hàng ngày VIC")
plt.xlabel("Lợi suất")
plt.ylabel("Tần suất")
plt.show()
```

```python
# Vẽ biểu đồ khối lượng giao dịch 
sns.lineplot(x="time", y="volume", data=df)
plt.title("Khối lượng giao dịch VIC")
plt.gcf().autofmt_xdate()
plt.tight_layout()
plt.show()
```

```python
# vẽ biểu đồ phân tán (scatter plot)
sns.scatterplot(x="volume", y="close", data=df)
plt.title("Quan hệ khối lượng và giá VIC")
plt.xlabel("Khối lượng")
plt.ylabel("Giá đóng cửa")
#plt.gcf().autofmt_xdate()
#plt.tight_layout()
plt.show()
```
