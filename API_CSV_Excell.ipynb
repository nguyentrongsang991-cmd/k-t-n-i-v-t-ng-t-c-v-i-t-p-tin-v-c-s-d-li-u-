## Thông tin cho nhiều cổ phiếu

```python
# Danh sách mã cổ phiếu cần lấy
tickers = ['AAPL', 'MSFT', 'TSLA', 'AMZN', 'GOOGL']
```

```python
# Tạo danh sách rỗng để lưu dữ liệu
summary_info = []
```

```python
# Vòng lặp qua từng mã cổ phiếu
for ticker in tickers:
    print(f"Đang lấy dữ liệu cho {ticker}...")
    stock = yf.Ticker(ticker)
    info = stock.info
    
    # Trích xuất các trường thông tin cần thiết
    company_name = info.get('longName', 'N/A')
    sector = info.get('sector', 'N/A')
    market_cap = info.get('marketCap', 'N/A')
    trailing_pe = info.get('trailingPE', 'N/A')
    forward_pe = info.get('forwardPE', 'N/A')
    roe = info.get('returnOnEquity', 'N/A')
    
    # Thêm vào danh sách
    summary_info.append({
        'Ticker': ticker,
        'Company Name': company_name,
        'Sector': sector,
        'Market Cap (USD)': market_cap,
        'Trailing P/E': trailing_pe,
        'Forward P/E': forward_pe,
        'Return on Equity (ROE)': roe
    })

print("\nĐã lấy xong tất cả thông tin!")
```

```python
summary_info
```

```python
# Chuyển danh sách thành DataFrame
df_summary_info = pd.DataFrame(summary_info)
```

```python
# Danh sách mã cổ phiếu cần lấy
tickers = ['AAPL', 'MSFT', 'TSLA', 'AMZN', 'GOOGL']
```

```python
# Tải dữ liệu và lưu từng file theo từng mã CP
for ticker in tickers:
    print(f"Đang tải dữ liệu cho: {ticker}")
    
    # Lấy dữ liệu từ Yahoo Finance
    data = yf.download(ticker, start='2023-01-01', end='2024-01-01')
    
    # Đặt tên file
    file_path = os.path.join(folder_name, f"{ticker}_historical_data.xlsx")
   
    # Lưu ra file Excel
    data.to_excel(file_path)
    
    print(f"Đã lưu {ticker} vào {file_path}")

print(f" \n Hoàn thành việc tải và lưu tất cả file cổ phiếu vào thư mục", folder_name)
```

# ## Lưu dữ liệu các CP vào 1 file mỗi CP là 1 sheet , bổ sung thông tin cơ bản

```python
# import các thư viện
import yfinance as yf
import pandas as pd
import os
```

```python
# Danh sách mã cổ phiếu cần lấy
tickers = ['AAPL', 'MSFT', 'TSLA', 'AMZN', 'GOOGL']
```

```python
# Tạo một dictionary để lưu dữ liệu
all_data = {}
```

```python
# Lấy dữ liệu từng mã cổ phiếu
for ticker in tickers:
    print(f"Đang tải dữ liệu cho {ticker}...")
    stock = yf.Ticker(ticker)
    
    # Lấy dữ liệu lịch sử giá
    hist_data = stock.history(start='2023-01-01', end='2024-01-01')
    
    # Xóa timezone
    hist_data.index = hist_data.index.tz_localize(None)
    
    # Lấy thông tin công ty
    info = stock.info
    company_name = info.get('longName', 'Unknown Company')
    sector = info.get('sector', 'Unknown Sector')
    
    # Thêm cột Company Name và Sector vào dữ liệu giá
    hist_data['Company Name'] = company_name
    hist_data['Sector'] = sector
    
    # Lưu vào dictionary
    all_data[ticker] = hist_data

print("\nĐã hoàn thành tải dữ liệu cho tất cả cổ phiếu!")
```

```python
# Lưu vào một file Excel, mỗi sheet là một mã cổ phiếu
output_file = "All_Stocks_Data.xlsx" # tạo biến tên là output_file để lưu trữ đường dẫn và tên file Excel đầu ra.
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    for ticker, data in all_data.items():
        # Giới hạn tên sheet <= 31 ký tự
        sheet_name = ticker if len(ticker) <= 31 else ticker[:31]
        data.to_excel(writer, sheet_name=sheet_name)
        
print(f"\nTất cả dữ liệu đã được lưu vào file Excel: {output_file}")
```
